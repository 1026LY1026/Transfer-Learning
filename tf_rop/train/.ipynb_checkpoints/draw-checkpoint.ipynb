```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from pylab import mpl

# 设置matplotlib的配置
# mpl.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体
mpl.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

# 忽略警告提示
import warnings

warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./XJ_well_data/well_3.csv')
```

```python
data
```

```python
data1=data[:int(len(data)/2)]
data2=data[int(len(data)/2):]
```

```python
plt.figure(figsize=(18, 6))
plt.plot(data1['MD'], data1['ROPA'], label='Well_3_1', color='blue', linewidth=1)
plt.plot(data2['MD'], data2['ROPA'], label='Well_3_2', color='green', linewidth=1)
plt.ylabel("rop", fontsize=18)
plt.xlabel('depth', fontsize=18)
plt.savefig('well3_1_2.png')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 假设data1和data2已经被定义，包含'MD'（深度）和'ROPA'（钻井速率）数据

# 设置图像大小
plt.figure(figsize=(18, 6))

# 定义偏重点的亮色调的颜色
colors = ['#FF6700', '#0077BE']

# 绘制数据曲线，添加标签
plt.plot(data1['MD'], data1['ROPA'], label='Well 3-1', color=colors[0], linewidth=1.5)
plt.plot(data2['MD'], data2['ROPA'], label='Well 3-2', color=colors[1], linewidth=1.5)

# 设置白色背景
plt.rcParams['axes.facecolor'] = 'white'

# 设置坐标轴标签和标题
plt.ylabel("Rate of Penetration (ROP)", fontsize=20)
plt.xlabel('Depth (ft)', fontsize=20)
plt.title('ROP vs. Depth for Two Wells', fontsize=24)

# 显示图例，设置图例位置在图表外部
plt.legend(fontsize=14, loc='upper left', bbox_to_anchor=(1,1))

# 调整坐标轴刻度和字体大小
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# 显示网格，设置网格颜色为亮色
plt.grid(True, which='both', linestyle='--', linewidth=1, color='lightgrey')

# 保存图像前调整布局
plt.tight_layout()

# 保存图像，设置背景为白色
plt.savefig('well3_1_2.png', dpi=300, facecolor='white')

# 显示图表
plt.show()
```
